```python
def is_prime(x):
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

def next_prime(n):
    i = n + 1
    while not is_prime(i):
        i += 1
    return i

print(next_prime(8))
print(next_prime(11))
```

```python
def prime_dist(start, dist):
    # p1, p2 = None, None
    if is_prime(start):
        p1 = start
    else:
        p1 = next_prime(start)

    p2 = next_prime(p1)

    while p2 < start * 10:
        if (p2 - p1) >= dist:
            return p1, p2

        p1 = p2
        p2 = next_prime(p1)

    return None

print(prime_dist(3, 6))
print(prime_dist(1000, 10))
print(prime_dist(1000, 20))
print(prime_dist(1000, 35))
print(prime_dist(1000, 37))
print(prime_dist(1000000, 100))
```

```python
g = ['..........', '.###.##...', '.....##...', '.......##.', '.......##.', '..........']

def living_neighbours(grid, row, col):
    count = 0
    if row + 1 <= len(grid) - 1 and grid[row + 1][col] == '#':
        count += 1

    if row - 1 >= 0 and grid[row - 1][col] == '#':
        count += 1

    if col + 1 <= len(grid[row]) - 1 and grid[row][col + 1] == '#':
        count += 1

    if col - 1 >= 0 and grid[row][col - 1] == '#':
        count += 1

    return count

print(living_neighbours(g, 1, 1))
print(living_neighbours(g, 1, 2))
print(living_neighbours(g, 3, 1))
```

```python
print(g)

s = g[::]

chars = [c for c in s[0]]
chars[3] = 'K'
s[0] = ''.join(chars)

print('-----')

print(g)
print(s)
```

```python
def state(grid):
    state = grid[::]

    for i in range(len(state)):
        chars = [c for c in grid[i]]
        for j in range(len(state[i])):
            n = living_neighbours(state, i, j)
            dead = state[i][j] == '.'

            if dead and n == 2:
                dead = False
            elif not dead and (n < 2 or n > 3):
                dead = True

            chars[j] = '.' if dead else '#'

        grid[i] = ''.join(chars)

g = ['..........', '.###.##...', '.....##...', '.......##.', '.......##.', '..........']

print(g)
state(g)
print(g)

for _ in range(1000):
    state(g)
    print(g)
```
